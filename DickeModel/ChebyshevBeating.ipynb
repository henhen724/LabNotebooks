```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-1., 1., 100)
plt.plot(x,np.log(1.+np.exp(x/0.02)))
plt.xlabel("T")
plt.ylabel("M")
plt.title("Order Parameter Elbow")
```

```python
x = np.linspace(-1., 1., 100)
plt.plot(x,1/0.02*np.exp(x/0.02)/(1.+np.exp(x/0.02)))
plt.xlabel("T")
plt.ylabel("M'(T)")
plt.title("Order Parameter First Derivative")
```

```python
x = np.linspace(-1., 1., 212)
plt.plot(x,((1/0.02)**2)*(np.exp(x/0.02)/(1.+np.exp(x/0.02)) - np.exp(2*x/0.02)/(1.+np.exp(x/0.02))**2))
plt.xlabel("T")
plt.ylabel("M''(T)")
plt.title("Order Parameter Second Derivative")
```

```python
x = np.linspace(-1., 1., 212)
plt.plot(x,((1/0.02)**3)*(np.exp(x/0.02)/(1.+np.exp(x/0.02)) - 3.*np.exp(2*x/0.02)/(1.+np.exp(x/0.02))**2 + 2.*np.exp(3*x/0.02)/(1.+np.exp(x/0.02))**3))
plt.xlabel("T")
plt.ylabel("M'''(T)")
plt.title("Order Parameter Third Derivative")
```
