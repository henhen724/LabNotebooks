```julia
using DifferentialEquations
using Plots
using LaTeXStrings
using JLD2
using Printf
```

```julia
# 1: (a+a^\dag)/2 
# 2: i(a-a^\dag)/2 
# 3: S_x 
# 4: S_y 
# 5: S_z
 = 1.0 # MHz
 = 1.0 # MHz
 = 0.01

g_c = 1/2*sqrt(*(^2 + ^2)/)

g = 2.0

angle = 2*pi*rand(Float64, 2)
angle[1] = angle[1] % pi
x = rand(Float64)
inital = [x,0., sin(angle[1]), cos(angle[1])*cos(angle[2]), cos(angle[1])*sin(angle[2])]


function dicke(du, u, p, t)
    du[1] = *u[2]-*u[1]
    du[2] = -*u[1]-*u[2]+g*u[3]
    du[3] = -*u[4]
    du[4] = *u[3]+g*u[1]*u[5]
    du[5] = -g*u[1]*u[4]
end

function _dicke(du, u, p, t)
    du[1] = -sqrt(/2)
    du[2] = -sqrt(/2)
    du[3] = 0.
    du[4] = 0.
    du[5] = 0.
end

prob_sde_lorenz = SDEProblem(dicke, _dicke, inital, (0.0, 4000.0))
sol = solve(prob_sde_lorenz, SKenCarp())
plot(sol, idxs = (1, 2))
```

```julia
plot(sol, idxs = (3, 4, 5))
```

```julia
twopointreal = [sum([sol[j][1]sol[j+i][1] for j=1:length(sol)-i])/(length(sol)-i) for i=1:100]
```

```julia
twopointimag = [sum([sol[j][2]sol[j+i][2] for j=1:length(sol)-i])/(length(sol)-i) for i=1:10000]
```

```julia
function save_correlators(, , , g)
    # 1: (a+a^\dag)/2 
    # 2: i(a-a^\dag)/2 
    # 3: S_x 
    # 4: S_y 
    # 5: S_z
    # = 1.0 # MHz
    # = 1.0 # MHz
    # = 0.01

    g_c = 1/2*sqrt(*(^2 + ^2)/)

    if g > g_c
        inital = [g*/(^2 + ^2), 0, 1.0, 0, (^2 + ^2)/*/g^2]
    else
        inital = [0., 0., 1.0, 0., 0.]
    end

    function dicke(du, u, p, t)
        du[1] = *u[2]-*u[1]
        du[2] = -*u[1]-*u[2]+g*u[3]
        du[3] = -*u[4]
        du[4] = *u[3]+g*u[1]*u[5]
        du[5] = -g*u[1]*u[4]
    end

    function _dicke(du, u, p, t)
        du[1] = -sqrt(/2)
        du[2] = -sqrt(/2)
        du[3] = 0.
        du[4] = 0.
        du[5] = 0.
    end

    prob_sde_lorenz = SDEProblem(dicke, _dicke, inital, (0.0, 4000.0))
    sol = solve(prob_sde_lorenz)

    twopointxx = [sum([sol[j][1]sol[j+i][1] for j=1:length(sol)-i])/(length(sol)-i) for i=1:10000]
    twopointxy = [sum([sol[j][1]sol[j+i][2] for j=1:length(sol)-i])/(length(sol)-i) for i=1:10000]
    twopointyy = [sum([sol[j][2]sol[j+i][2] for j=1:length(sol)-i])/(length(sol)-i) for i=1:10000]
    
    JLD2.save_object("DickeRuns/d=$ w=$ k=$ g=$g xx.jld2", twopointxx)
    JLD2.save_object("DickeRuns/d=$ w=$ k=$ g=$g xy.jld2", twopointxy)
    JLD2.save_object("DickeRuns/d=$ w=$ k=$ g=$g yy.jld2", twopointyy)
end
```

```julia
save_correlators(1.0, 1.0, 0.01, 0.01)
```

```julia
# 1: (a+a^\dag)/2 
# 2: i(a-a^\dag)/2 
# 3: S_x 
# 4: S_y 
# 5: S_z
# 6: g(t)
# 7: t of last bang
 = 1.0 # MHz
 = 1.0 # MHz
 = 0.01
g_init = 0.5

g_c = 1/2*sqrt(*(^2 + ^2)/)

# if g > g_c
#     inital = [g*/(^2 + ^2), 0, 1.0, 0, (^2 + ^2)/*/g^2]
# else
#     inital = [0., 0., 1.0, 0., 0.]
# end
angle = 2*pi*rand(Float64, 2)
angle[1] = angle[1] % pi
x = rand(Float64)
inital = [x,0., sin(angle[1]), cos(angle[1])*cos(angle[2]), cos(angle[1])*sin(angle[2]), g_init, 0.0]

t_last_bang = 0.0
cool_down_t = 10.0
max_x = 0.1

function dicke(du, u, p, t)
    if abs(u[1]) < max_x && t - u[7] > cool_down_t
        u[7] = t
        u[6] = -u[6]
    end

    du[1] = *u[2]-*u[1]
    du[2] = -*u[1]-*u[2]+g*u[3]
    du[3] = -*u[4]
    du[4] = *u[3]+g*u[1]*u[5]
    du[5] = -g*u[1]*u[4]
    du[6] = 0.0
    du[7] = 0.0
end

function _dicke(du, u, p, t)
    du[1] = -sqrt(/2)
    du[2] = -sqrt(/2)
    du[3] = 0.
    du[4] = 0.
    du[5] = 0.
    du[6] = 0.0
    du[7] = 0.0
end

prob_sde_lorenz = SDEProblem(dicke, _dicke, inital, (0.0, 4000.0))
sol = solve(prob_sde_lorenz, SKenCarp())
```
