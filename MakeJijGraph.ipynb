```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.cm as cm
```

```python
x = np.linspace(-100.,100.,500)
y = np.linspace(-100.,100.,500)
X,Y=np.meshgrid(x,y)
w_0 = 35.
xi = 5.
X_0 = 50
Y_0 = 10
Z = 1/np.pi*np.cos(2*(X*X_0+Y*Y_0)/w_0**2)\
+np.exp(-np.sqrt((X-X_0)**2+(Y-Y_0)**2)/xi)/np.sqrt(np.sqrt((X-X_0)**2+(Y-Y_0)**2)/xi)\
+np.exp(-np.sqrt((X+X_0)**2+(Y+Y_0)**2)/xi)/np.sqrt(np.sqrt((X+X_0)**2+(Y+Y_0)**2)/xi)
```

```python
plt.pcolor(X,Y,Z, vmin=-0.5, vmax=0.5, cmap="RdBu")
plt.colorbar()
plt.xlabel("Position ($\mu$m)", fontsize=28)
plt.ylabel("Position ($\mu$m)", fontsize=28)
plt.xticks(np.arange(-100,150,step=50), fontsize=16)
plt.yticks(np.arange(-100,150,step=50), fontsize=16)
```
